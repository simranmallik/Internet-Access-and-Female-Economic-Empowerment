### Finding Common Countries Among Datasets by Year

```python
# Get only individual countries by filtering out regional aggregates
# World Bank uses 2-3 letter codes for countries; regional aggregates have codes like 'EAS', 'SSF', etc.
# We'll use a list of valid ISO country codes to filter

import pycountry
valid_country_codes = {country.alpha_3 for country in pycountry.countries}

def filter_countries_only(df):
    return df[df['Country Code'].isin(valid_country_codes)]

internet_countries = filter_countries_only(internet_df)
# female_internet_past3_countries = filter_countries_only(female_internet_past3_df)
female_emp_upper_countries = filter_countries_only(female_emp_upper_positions_df)
female_emp_countries = filter_countries_only(female_emp_df)

# For each year between 2015-2019, find countries that have values in ALL 3 datasets
results = []

for year in range(2015, 2026):
    internet_year = set(
        internet_countries[
            (internet_countries['Year'] == year) &
            (internet_countries['Value'].notna()) &
            (internet_countries['Value'] != 0)
        ]['Country Code']
    )

    emp_upper_year = set(
        female_emp_upper_countries[
            (female_emp_upper_countries['Year'] == year) &
            (female_emp_upper_countries['Value'].notna()) &
            (female_emp_upper_countries['Value'] != 0)
        ]['Country Code']
    )

    emp_year = set(
        female_emp_countries[
            (female_emp_countries['Year'] == year) &
            (female_emp_countries['Value'].notna()) &
            (female_emp_countries['Value'] != 0)
        ]['Country Code']
    )

    # Countries present in ALL 3 datasets for this year
    common_countries = internet_year & emp_upper_year & emp_year

    results.append({
        'Year': year,
        'Internet': len(internet_year),
        'Female Employment': len(emp_year),
        'Female Mgmt Employment': len(emp_upper_year),
        'All 3 Datasets': len(common_countries)
    })

results_df = pd.DataFrame(results).set_index('Year')
print(results_df)
```

```python
# Get countries present in all 3 datasets for 2022
internet_2022 = set(
    internet_countries[
        (internet_countries['Year'] == 2022) &
        (internet_countries['Value'].notna()) &
        (internet_countries['Value'] != 0)
    ]['Country Code']
)

emp_upper_2022 = set(
    female_emp_upper_countries[
        (female_emp_upper_countries['Year'] == 2022) &
        (female_emp_upper_countries['Value'].notna()) &
        (female_emp_upper_countries['Value'] != 0)
    ]['Country Code']
)

emp_2022 = set(
    female_emp_countries[
        (female_emp_countries['Year'] == 2022) &
        (female_emp_countries['Value'].notna()) &
        (female_emp_countries['Value'] != 0)
    ]['Country Code']
)

common_2022 = internet_2022 & emp_upper_2022 & emp_2022

# Print as country names instead of codes
code_to_name = dict(zip(internet_df['Country Code'], internet_df['Country Name']))
common_2022_names = sorted([code_to_name.get(code, code) for code in common_2022])

# print(f"Number of countries in all 3 datasets for 2022: {len(common_2022_names)}\n")
# for country in common_2022_names:
#     print(country)
```

#### Subset the datasets so they contain the 92 common countries and information for 2022 only

```python
##Subset for countries
#print(type(common_2022_names))
internet_countries_2022 = internet_countries[(internet_countries["Country Name"].isin(common_2022_names)) & (internet_countries["Year"] == 2022)]
internet_countries_2022["Disaggregation"].value_counts()
##We are looking at total population internet use
internet_countries_2022 = internet_countries_2022[internet_countries_2022["Disaggregation"] == "total"]
internet_countries_2022["Indicator Name"] = "internet use (total)"
internet_countries_2022 #We have 92 observations
#internet_countries
```

```python
empUpper_2022 = female_emp_upper_countries[(female_emp_upper_countries["Country Name"].isin(common_2022_names)) & (female_emp_upper_countries["Year"] == 2022)]
empUpper_2022 #We have 92 observations
empUpper_2022["Indicator Name"] = "female upper management"
empUpper_2022
#empUpper_2022["Disaggregation"].value_counts() #We have 92 observations
```

```python
fem_emp_2022 = female_emp_countries[(female_emp_countries["Country Name"].isin(common_2022_names)) & (female_emp_countries["Year"] == 2022)]
fem_emp_2022["Indicator Name"] = "female employment"
fem_emp_2022["Disaggregation"].value_counts() #We have 92 observations
fem_emp_2022
#female_emp_countries
```

### Concatenate Datasets

```python
#Concat all the data together
final_data = pd.concat([internet_countries_2022, empUpper_2022, fem_emp_2022])
#Remove redunant variables and year (since all data is from the same year it wont add significance to tests)
final_data.drop(columns= ["Year", "Indicator Code", "Country Code", "Disaggregation"], inplace = True)
##Look at the final dataset
#final_data["Country Name"].value_counts()


##Checking for empty strings in data
#empty_strings = (final_data['Value'] == "").sum()
#empty_strings ##No empty strings

final_data['Value'] = pd.to_numeric(final_data['Value'], errors='coerce')
final_data = final_data.dropna(subset=['Value'])
final_data ##No NaNs we can proceed with tests
```

#### Turn the dataframe into a pivot data for use in nonparametric tests

```python
#Turn data to a pivot table/wide format
final_table = final_data.pivot(index = "Country Name", columns = "Indicator Name", values = "Value")
final_table
```

### Running the Shapiro-Wilk Test for Normality
This test is applied to each of the columns to ensure the distributions of the data are non-normal for use of nonparameteric tests. However, the Shapiro-Wilk Test can lead to false positives due to its sensitvity to outliers so a histogram/Q-Q plot will be used to validate these results.

```python
import scipy.stats as stats
#Run Shapiro Wilks test for normality
stat, p = stats.shapiro(final_table["female employment"].dropna())
print("P-value:", p)
#We reject the null hypothesis
#Female employment data does not have a normal distribution
```

```python
#Run Shapiro Wilks test for normality
stat, p = stats.shapiro(final_table["female upper management"].dropna())
print("P-value:", p)
#We fail to reject null
#Female employment data might have a normal distribution
```

```python
#Run Shapiro Wilks test for normality
stat, p = stats.shapiro(final_table["internet use (total)"].dropna())
print("P-value:", p)
#We reject the null hypothesis
#Internet use does not have a normal distribution
```
